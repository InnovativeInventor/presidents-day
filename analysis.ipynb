```python
import csv
import numpy as np
import datetime
import calendar
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
days_of_week_total = []
days_of_week_curtis = []
with open("historical-days.csv", "r") as f:
    for count, each_event in enumerate(csv.DictReader(f)):
        if each_event["Head of School"] == "":
            break
        if "/" in each_event["Actual Day"]:
            date_list = each_event["Actual Day"].split("/")
            date = datetime.datetime(int("20" + date_list[2]), int(date_list[0]), int(date_list[1]))
            days_of_week_total.append(date.weekday())
            if each_event["Head of School"] == "Curtis":
                days_of_week_curtis.append(date.weekday())
            print(calendar.day_name[date.weekday()])
            print(date)
```

```python
sns.distplot(days_of_week_total, bins=5, kde=False)
plt.grid()
plt.gca().set(title='Cumulative Frequency of President\'s Day by Weekday', ylabel='Frequency', xlabel="Days of the Week");
```

```python
sns.distplot(days_of_week_curtis, bins=5, kde=False)
plt.grid()
plt.gca().set(title='Frequency of President\'s Day by Weekday under Dr. Curtis', ylabel='Frequency', xlabel="Days of the Week");
```
